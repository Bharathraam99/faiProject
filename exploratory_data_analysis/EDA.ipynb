```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
stock_names = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NVDA", "META", "NFLX", "ADBE", "INTC", 
    "BRK.B", "JNJ", "PG", "XOM", "V", "KO", "DIS", "MA", "AMD", "CVX"]
datasets_path = f"../new_dataset/historical_data/"
datasets = []
```

```python
for stock in stock_names:
    file_path = f"{datasets_path}{stock}.csv" 
    try:
        data = pd.read_csv(file_path)
        datasets.append(data)
        print(f"Data for {stock} loaded successfully.")  
    except FileNotFoundError:
        print(f"File not found for stock: {stock} at path: {file_path}")
```

## EDA

```python
# Check for missing values in each stock dataset
for i, stock in enumerate(stock_names):
    print(f"Missing values for {stock}:")
    print(datasets[i].isnull().sum())
    print("\n")
```

```python
# Display summary statistics for each stock's numerical columns
for i, stock in enumerate(stock_names):
    print(f"Summary statistics for {stock}:")
    print(datasets[i].describe())
    print("\n")
```

## Daily Returns: The graph shows daily returns for each stock, which indicates the percentage change in price from the previous day. Positive values mean the stock price increased, and negative values mean it decreased.

```python
import matplotlib.pyplot as plt

# Calculate daily returns and store in a dictionary
daily_returns = {}
for i, stock in enumerate(stock_names):
    if 'Close' in datasets[i].columns:
        datasets[i]['Daily Return'] = datasets[i]['Close'].pct_change()  # Calculate daily return
        daily_returns[stock] = datasets[i]['Daily Return']  # Store daily return

# Plot daily returns for each stock in separate graphs
for stock in stock_names:
    plt.figure(figsize=(10, 4))
    plt.plot(datasets[stock_names.index(stock)]['Date'], daily_returns[stock], label=f"{stock} Daily Return")
    plt.title(f"Daily Returns for {stock}")
    plt.xlabel("Date")
    plt.ylabel("Daily Return")
    plt.legend()
    plt.show()
```

## The daily returns plots show that Bitcoin and Tesla are the most volatile, with frequent large swings, while Gold is the most stable, with minimal fluctuation. Most tech stocks (like Apple, Microsoft, and NVIDIA) have moderate volatility, generally centered around zero but with occasional spikes in response to market news. This helps identify which stocks are riskier (like Tesla and Bitcoin) and which are more stable (like Gold), making it easier to balance risk when building a portfolio.

## Cumulative Returns Plot: Shows how much each stock has grown over time.

```python
# Plot cumulative returns for each stock
for stock in stock_names:
    plt.figure(figsize=(10, 4))
    data = datasets[stock_names.index(stock)]
    if 'Close' in data.columns:
        data['Cumulative Return'] = (1 + data['Daily Return']).cumprod() - 1  # Calculate cumulative returns
        plt.plot(data['Date'], data['Cumulative Return'], label=f"{stock} Cumulative Return")
        plt.title(f"Cumulative Returns for {stock}")
        plt.xlabel("Date")
        plt.ylabel("Cumulative Return")
        plt.legend()
        plt.show()
```

## The cumulative returns plots show that tech stocks (like Apple, Microsoft, and NVIDIA) and Tesla have strong upward growth trends, indicating high returns. Bitcoin exhibits explosive growth but also extreme volatility, with sharp rises and drops. Gold is more stable with moderate growth, while the Treasury Bill (IRX) has a consistent negative trend. Overall, tech stocks and Bitcoin offer high growth potential with high risk, while Gold is more stable but with lower returns.

```python
import pandas as pd

# Calculate daily returns and store them in a dictionary
daily_returns = {}
for i, stock in enumerate(stock_names):
    if 'Close' in datasets[i].columns:
        datasets[i]['Daily Return'] = datasets[i]['Close'].pct_change()
        daily_returns[stock] = datasets[i]['Daily Return']

daily_returns_df = pd.DataFrame(daily_returns)
daily_returns_df.dropna(inplace=True)
```

```python
# Calculate expected returns (mean daily return * trading days) and volatility (std of daily returns)
expected_returns = daily_returns_df.mean() * 252  # Annualized expected return
volatility = daily_returns_df.std() * (252 ** 0.5)  # Annualized volatility
metrics_df = pd.DataFrame({'Expected Return': expected_returns, 'Volatility': volatility})
print(metrics_df)
```

## We can use the above observation in Genetic Algorithm for portfolio optimization by getting to know which assets offer high returns with high risk (e.g., Tesla, Bitcoin), stable growth (e.g., Apple, Microsoft), and low risk (e.g., Gold). The algorithm can balance assets based on desired risk tolerance and return goals, leveraging high-return assets with high volatility alongside more stable ones.

```python
# Calculate the correlation matrix of daily returns
correlation_matrix = daily_returns_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix of Daily Returns")
plt.show()
```

```python
# Calculate the covariance matrix of daily returns
covariance_matrix = daily_returns_df.cov() * 252  # Annualized covariance matrix
print(covariance_matrix)
```
